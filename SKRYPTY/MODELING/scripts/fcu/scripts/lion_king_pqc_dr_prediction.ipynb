### Preparing dataset for sending to model

```python
final_wf_dr_score = final_wf_dr[[
    'ScreenedParties', 'OwnershipLayers', 'Experience_proj', 'Experience_team', 'Hour_numeric',
    'Group_cases', 'NO_errors_last_5_days_DR', 'NO_cases_last_5_days_DR', 'NO_errors_last_month_DR',
    'NO_cases_month_DR','PartyType', 'TLAssignedName', 'C DDRiskLevel', 'ProcessingUnit', 'ESR', 'CRS',
    'FATCA', 'GreenlightedDR', 'Day', 'Population_match']]
```

```python
final_wf_dr_score.shape
```

```python
final_wf_dr_score.columns = [
    'ScreenedParties', 'OwnershipLayers', 'Experience_proj', 'Experience_team', 'Hour_numeric',
    'Group_cases', 'NO_errors_last_5_days', 'NO_cases_last_5_days', 'NO_errors_last_month',
    'NO_cases_month','PartyType', 'TLAssigned', 'CDDRiskLevel', 'ProcessingUnit', 'ESR', 'CRS',
    'FATCA','GreenlightedDR', 'Day', 'Population_match']
```

```python
final_wf_dr_score
```

```python
# Fixing TL Assigned column
final_wf_dr_score.TLAssigned = final_wf_dr_score.TLAssigned.str.split(',').str[0]
```

```python
final_wf_dr_score
```

```python
# Selecting columns to be converted into dummy variables
final_wf_dr_dummies = pd.get_dummies(final_wf_dr_score,
                                     columns=['TLAssigned', 'PartyType', 'CDDRiskLevel', 'ProcessingUnit', 'ESR',
                                              'CRS', 'FATCA', 'GreenlightedDR', 'Day', 'Population_match'])
```

```python
final_wf_dr_dummies.shape
```

```python
# Generating empty train df
train = pd.DataFrame(columns=[
    'ScreenedParties', 'OwnershipLayers', 'Experience_proj', 'Experience_team', 'Hour_numeric',
    'Group_cases', 'NO_errors_last_5_days', 'NO_cases_last_5_days', 'NO_errors_last_month',
    'NO_cases_month', 'PartyType_Subsidiary', 'TLAssigned_Makowska', 'TLAssigned_Kolodziejczyk',
    'TLAssigned_Jaszewski', 'TLAssigned_Michalik', 'TLAssigned_Armannsson', 'TLAssigned_Rybka',
    'TLAssigned_Galtie', 'TLAssigned_Skrzynecki', 'TLAssigned_Holowacz', 'TLAssigned_Gronowski',
    'CDDRiskLevel_Normal', 'CDDRiskLevel_Low', 'CDDRiskLevel_Increased', 'ProcessingUnit_Gdansk',
    'ProcessingUnit_MidCorp', 'ProcessingUnit_DIBA', 'ProcessingUnit_Rotterdam', 'ProcessingUnit_Not_assigned',
    'ProcessingUnit_Midcorp', 'ESR_No_ESR_needed', 'ESR_ESR_linked', 'CRS_FALSE', 'FATCA_FALSE',
    'GreenlightedDR_FALSE', 'Day_Wednesday', 'Day_Friday', 'Day_Tuesday', 'Day_Monday', 'Population_match_FALSE'])
```

```python
# Finding misssing cols on datasets to be scored
missing_cols = set(train.columns) - set(final_wf_dr_dummies.columns)
```

```python
missing_cols
```

```python
# Adding zero to missing cols
for c in missing_cols:
    final_wf_dr_dummies[c] = 0
    
final_wf_dr_dummies_2 = final_wf_dr_dummies[train.columns]    
```

```python
# Fill NA cols with zero
final_wf_dr_dummies_2 = final_wf_dr_dummies_2.fillna(0)
```

```python
# Define numeric cols
num_cols = [
    'ScreenedParties', 'OwnershipLayers', 'Experience_proj', 'Experience_team', 'Hour_numeric', 'Group_cases',
    'NO_errors_last_5_days', 'NO_cases_last_5_days', 'NO_errors_last_month', 'NO_cases_month']
```

```python
# Normalizing cols
scaler = StandardScaler()
final_wf_dr_dummies_2[num_cols] = scaler.fit_transform(final_wf_dr_dummies_2[num_cols])
```

### Running model for prediction

```python
ynew = model.predict_proba(final_wf_dr_dummies_2)
```

```python
pqc = pd.DataFrame(ynew, columns=['pqc'])
```

```python
scored_wf = pd.concat([final_wf_dr,pqc],axis=1)
```
